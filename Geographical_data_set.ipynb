```python
import pandas as pd
import numpy as np
```

```python
# Generate random geographical data
data = {
    'longitude': np.random.uniform(-180, 180, size=100),
    'latitude': np.random.uniform(-90, 90, size=100),
    'temperature': np.random.normal(15, 5, size=100),
    'elevation': np.random.exponential(1000, size=100),
    'area': np.random.uniform(1, 1000, size=100)
}
```

```python
# Create a DataFrame
df = pd.DataFrame(data)
df.head(50)
```

```python
df.to_excel('geographical_data.xlsx')
```
